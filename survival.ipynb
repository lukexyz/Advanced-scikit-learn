```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Monthly Survival Rate

cohorts = [0.97, 0.95, 0.92]
months = 12

v1 = [x*cohorts[0]**(i) for i, x in enumerate(np.array([1.]*months))]
v2 = [x*cohorts[1]**(i) for i, x in enumerate(np.array([1.]*months))]
v3 = [x*cohorts[2]**(i) for i, x in enumerate(np.array([1.]*months))]

df = pd.DataFrame({'v1': v1, 'v2': v2, 'v3': v3})
df.plot()
```

```python
# Avg. Monthly profit

m_profit = 10

print('Expected Annual Profit')
for v in df[['v1', 'v2', 'v3']]:
    print(f'{v}\t £{ (df[v]*m_profit).sum():0.2f}')
    
df['v1_profit'] = (df['v1']*m_profit).cumsum().round(2)
df['v2_profit'] = (df['v2']*m_profit).cumsum().round(2)
df['v3_profit'] = (df['v3']*m_profit).cumsum().round(2)

df[['v1_profit','v2_profit','v3_profit']].plot()
```
